```python
from imu.snowflake import SnowflakeClient, SnowflakeConfig
```

```python
# All the snow flake interaction us a package called IBIS 
client = SnowflakeClient()
client
```

```python
## QUERY SNOWFLAKE FOR THE SAMPLE TRACKING SHEET WHICH CONTAINS THE S3 LOCATIONS FOR THE FCS FILES ##
samples_table = client.table('V_FCS_TRACKED_SAMPLES', database='IMU.IMU')
samples_table
```

```python
# execute takes the data and returns a df
df = samples_table.limit(5).execute()
df
```

```python
# filter on rows 
samples_table.filter(samples_table['Panel'] == 'Tsv4').limit(5).execute()
```

```python
samples_table.filter(samples_table['Panel'] == 'Tsv4',samples_table['Project'] == 'BHAM').limit(10).execute()
```

```python
# without execute 
bham_samples = samples_table.filter(samples_table['Project'] == 'BHAM')
bham_samples
```

```python
# select you can you to list column you want to take 
bham_parquets = bham_samples.select(
    bham_samples['Sample ID'],
    bham_samples['OUTPUT_RAW_PARQUET_KEY'], 
    bham_samples['OUTPUT_XFORM_PARQUET_KEY'],
    bham_samples['Panel']
).execute()
bham_parquets
```

```python
# Opening parquets as a dateframe
import pyarrow.parquet as papq
import pandas as pd

file = bham_parquets.loc[0,'OUTPUT_RAW_PARQUET_KEY']

f = papq.ParquetFile(f"s3://{file}")
f.read().to_pandas()
```
